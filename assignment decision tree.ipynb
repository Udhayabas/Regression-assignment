```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv") 
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independant=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independant
```

```python
dependant=dataset[[  'charges']]
```

```python
dependant
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor= DecisionTreeRegressor(criterion='mse',splitter='best',max_features='log2')
regressor=regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
